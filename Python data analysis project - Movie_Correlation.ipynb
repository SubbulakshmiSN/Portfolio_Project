## Movie Correlation using matplot & seaborn

```python
#importing data
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams["figure.figsize"]=(12,8)

#read in the data
df=pd.read_csv(r"C:\Users\Admin\Desktop\Python\Csv files\movies.csv")
df.head()
```

```python
#cleaning the data
for col in df.columns:
    percent_missing=np.mean(df[col].isnull())
    print("{} - {}%".format(col, percent_missing))
```

```python
df=df.dropna()
df
```

```python
df["budget"]
```

```python
#changing data types
df["budget"]=df["budget"].astype("int64")
df["gross"]=df["gross"].astype("int64")
df.head(5)
```

```python
df["Corrected_Release_Year"]=df["released"].str.extract(pat="([0-9]{4})").astype("int")
df.head()
```

```python
df.sort_values("gross", ascending= False, inplace = True)
```

```python
#df.drop_duplicates()
```

```python
#Budget have high correlation
# company high correlation
```

```python
df.head()
```

```python
# lets build a scatter plot with budget vs gross

plt.scatter(x=df["budget"], y=df["gross"])
plt.title("Budget vs Gross Earnings")
plt.xlabel("Budget Earnings")
plt.ylabel("Gross Earnings")

plt.show()

```

```python
#to determine these are correlated using regression 
#plot budget vs gross using seaborn
sns.regplot(x="budget", y="gross", data=df, scatter_kws= {"color":"red"}, line_kws={"color":"green"})
```

```python
# lets dig into correlation
df.corr(method="pearson")  #Types of correlation= "pearson", "kendall","spearman" by default its pearson
```

```python
df.corr(method="kendall")
```

```python
df.corr(method='spearman')
```

```python
# conclusion-High correlation between budget and gross is correct
```

```python
correlation_matrix=df.corr(method="pearson")
sns.heatmap(correlation_matrix, annot= True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
#now digging abt company 
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=="object"):
        df_numerized[col_name]=df_numerized[col_name].astype("category")
        df_numerized[col_name]=df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix=df_numerized.corr(method="pearson")
sns.heatmap(correlation_matrix, annot= True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
#to unstack
df_numerized.corr()
```

```python
correlation_mat= df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs= corr_pairs.sort_values()
sorted_pairs
```

```python
high_correlation=sorted_pairs[(sorted_pairs)>0.5]
high_correlation
```
